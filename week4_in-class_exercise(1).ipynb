```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Read the data (from week3 exercise)

The dataset, same as week3, is attached in this repository.

```python
survey_data_df = pd.read_csv('pre-course_survey.csv')
```

## 2. Preprocess the data (from week3 exercise)

1. Rename all columns
2. Drop nan
3. Drop duplicates

```python
def process_data(df):
    # rename the columns
    old_cols = df.columns
    new_cols = ["timestamp", "python", "datascience", "git", "jupyter", "reproducibility", "nlp", "socialmedia_data", "goals", "skills", "concerns"]
    cols_dict = dict(zip(old_cols, new_cols))
    df.rename(columns=cols_dict, inplace=True)
    # drop nan
    df = df.dropna()
    # drop duplicates
    df = df.drop_duplicates()
    return df
```

```python
processed_data_df = process_data(survey_data_df)
```

### Line Plot

```python
processed_data_df = processed_data_df.replace({'Yes':1,'No':0})

processed_data_df['timestamp_datetime'] = pd.to_datetime(processed_data_df['timestamp'])
timestamp_grouped = processed_data_df.groupby(processed_data_df['timestamp_datetime'].dt.date)['socialmedia_data'].sum().reset_index()
# source: https://www.statology.org/pandas-group-by-month/
```

```python
x = timestamp_grouped['timestamp_datetime']
y = timestamp_grouped['socialmedia_data']
plt.plot(x, y)
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Daily Count of Students Responded "Yes" for Social Media Data Experience')
plt.yticks(np.arange(0, 6, 1))
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

### Scatter Plot

```python
plt.scatter(x, y)
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Daily Count of Students Responded "Yes" for Social Media Data Experience')
plt.yticks(np.arange(0, 6, 1))
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

###  Bar Plot

```python
py_freq_df = processed_data_df.value_counts(processed_data_df.python).reset_index()
py_freq_df.columns = ['python', 'frequency']

plt.bar(py_freq_df["python"], py_freq_df["frequency"])
plt.title("Distribution of Python Skill")
plt.xlabel("Python Skill")
plt.ylabel("Count")
plt.xticks(np.arange(1, 6, step=1))
plt.xlim([0, 6])
plt.show()
```
